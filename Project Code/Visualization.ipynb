```python
import pandas as pd
import numpy
import re
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
from os import path
from PIL import Image
from wordcloud import WordCloud
from wordcloud import STOPWORDS
from PIL import Image
import numpy as np
import warnings
warnings.filterwarnings("ignore")
#To plot figs on jupyter
%matplotlib inline
# figure size in inches
rcParams['figure.figsize'] = 8,6
```

```python
df = pd.read_csv('ProcessedData.tsv', delimiter = '\t', encoding = 'utf-8')
```

```python
b_news = df.loc[df['NewCategory'] == 'Business & Finance']
e_news = df.loc[df['NewCategory'] == 'Entertainment']
p_news = df.loc[df['NewCategory'] == 'Politics']
t_news = df.loc[df['NewCategory'] == 'Technology & Health']
c_news = df.loc[df['NewCategory'] == 'Crime']


b_title = b_news['title'].to_string()
e_title = e_news['title'].to_string()
p_title = p_news['title'].to_string()
t_title = t_news['title'].to_string()
c_title = c_news['title'].to_string()


b_image = np.array(Image.open("business.jpg"))
e_image = np.array(Image.open("Music.jpg"))
t_image = np.array(Image.open("health.jpg"))
p_image = np.array(Image.open("politics.jpg"))
c_image = np.array(Image.open("law.jpg"))


fig = plt.figure(figsize=(15,15))

# setting stop-words, so words like "the" and "it" are ignored
stopwords = set(STOPWORDS)

# business news cloud
ax1 = fig.add_subplot(321)
b_wordcloud = WordCloud(background_color='white', mask=b_image, collocations=False, stopwords=stopwords).generate(b_title)
ax1.imshow(b_wordcloud, interpolation='bilinear')
ax1.set_title('Business & Finance', size=20)
ax1.axis('off')


ax2 = fig.add_subplot(322)
e_wordcloud = WordCloud(background_color='white',mask=e_image, collocations=False, stopwords=stopwords).generate(e_title)
ax2.imshow(e_wordcloud, interpolation='bilinear')
ax2.set_title('Entertainment', size=20)
ax2.axis('off')

ax3 = fig.add_subplot(323)
t_wordcloud = WordCloud(background_color='white',mask=t_image, collocations=False, stopwords=stopwords).generate(t_title)
ax3.imshow(t_wordcloud, interpolation='bilinear')
ax3.set_title('Technology & Health', size=20)
ax3.axis('off')

ax4 = fig.add_subplot(324)
p_wordcloud = WordCloud(background_color='white',mask=p_image, collocations=False, stopwords=stopwords).generate(p_title)
ax4.imshow(p_wordcloud, interpolation='bilinear')
ax4.set_title('Politics', size=20)
ax4.axis('off')

ax5 = fig.add_subplot(325)
c_wordcloud = WordCloud(background_color='white',mask=c_image, collocations=False, stopwords=stopwords).generate(c_title)
ax5.imshow(c_wordcloud, interpolation='bilinear')
ax5.set_title('Crime', size=20)
ax5.axis('off')


fig.tight_layout()
```
